# DataGenerator test

```python
from spinesTS.data import DataGenerator, DataReader
from spinesTS.plotting import plot1d
```

```python
dg = DataGenerator()
```

## trigonometry_ds

```python
a = dg.trigonometry_ds(size=100, sin_cos_noise_fact=(0.4, 0.4, 0.2), random_state=0)
plot1d(a.data, figsize=(10, 6))
```

## white_noise

```python
b = dg.white_noise(100, random_state=0)
plot1d(b.data, figsize=(10, 6))
```

## random_walk

```python
c = dg.random_walk(size=100, started_zero=True, random_state=0)
plot1d(c.data, figsize=(10, 6))
```
